### Slicing dataframe

      * iloc and loc

```python
# Slicing rows adn columns 
movies_df.loc[5:21, ]
```

```python
movies_df.loc[5:21, ['Film', 'Lead Studio', 'Audience score %', 'Rotten Tomatoes %', 'Year']]
```

```python
# iloc: this uses index values to filter data
# start index: end index
movies_df.iloc[10:21]
```

```python
# start value  = 1 and end value is n -1
movies_df.iloc[10:21, 0:2]
```

```python
#Backward indexing  = Selecting last 4 columns
movies_df.iloc[10:21, -4:]
```

```python
# Filter teh columns by comparison
movies_df[movies_df['Year'] == 2011]
```

```python
# AND condition 

movies_df[(movies_df['Year'] == 2009) & (movies_df['Genre'] == 'Drama')]
```

```python
# OR condition

movies_df[(movies_df['Year'] == 2009) | (movies_df['Genre'] == 'Drama')]
```

```python
# 
movies_df[(movies_df['Audience score %'] >= 70) & (movies_df['Genre'] == 'Comedy')]
```

```python
# Finding duplicate rows
movies_df[movies_df.duplicated()]
```

```python
# Duplicated columns
movies_df.columns.duplicated()
```

```python
# Using filter method , selecting columns with values with 'a' in it
movies_df.filter(like='a').head()
```

```python
movies_df.filter(like='A').head()
```

```python
# Replace '$' from a column value 

movies_df['Worldwide Gross'] = movies_df['Worldwide Gross'].str.replace("$", "")
```

```python
movies_df
```

```python
# Convert values to lower case

movies_df['Genre'] = movies_df['Genre'].str.lower()
```

```python
# Typecasting a column value
movies_df['Worldwide Gross'] = movies_df['Worldwide Gross'].astype('float')
movies_df
```

```python
movies_df['Worldwide Gross'].dtype
```

```python
# Query 
movies_df.query("Genre == 'romance' and Profitability > 3")
```

```python
movies_df.query("`Lead Studio` in ('Independent', 'Summit')")
```

```python
# Genres not equals to romance
movies_df.query("Genre != 'romance'")
```
